```python
import urllib
import urllib.request
from bs4 import BeautifulSoup as BS
import os
```

```python
def makesoup(url):
    page = urllib.request.urlopen(url)
    soupdata = BS(page,"html.parser")
    return soupdata
```

```python
name = []
namesoup = makesoup("https://coinmarketcap.com/coins/views/all/")
for record in namesoup.findAll('tr'):
    if record.get('id'):
        name.append(record.get('id')[3:])

name
```

```python
datasaved = ""
file = open(os.path.expanduser("cryptodata.csv"),"wb")
for i in name:
    datasaved = ""
    soup = makesoup("https://coinmarketcap.com/currencies/"+str(i)+"/historical-data/?start=20130428&end=20180512")
    for record in soup.findAll('tr'):
        row = ""
        for data in record.findAll('td'):
            row = row + "|" + data.text
        datasaved = datasaved + "\n" + row[1:]+"|"+str(i)
    file.write(bytes(datasaved,encoding = 'ascii',errors='ignore'))
    print(i)
```

```python
file.close()
```
